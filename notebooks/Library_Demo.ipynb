# Hyperplane Arrangement Library Demo
This notebook demonstrates how to use the built-in arrangement presets from `hyperplane_arrangements.library`.

```
import sys
sys.path.append('../src')
from hyperplane_arrangements.library import *
from hyperplane_arrangements.arrangement import HyperplaneArrangement
```

## 1. Coxeter Arrangements
Coxeter arrangements of classical types: Braid (Type A), Type B/C, and Type D.

```
# Braid arrangement (Type A_{n-1})
A = braid(3)
print(f"Braid(3): {A.num_planes} planes in R^{A.n}")
print("Characteristic Polynomial:", A.characteristic_polynomial())

# Type B (same as Type C)
B = type_B(3)
print(f"\nType_B(3): {B.num_planes} planes in R^{B.n}")
print("Characteristic Polynomial:", B.characteristic_polynomial())
```

```
# Type D
D = type_D(3)
print(f"Type_D(3): {D.num_planes} planes in R^{D.n}")
print("Characteristic Polynomial:", D.characteristic_polynomial())
```

## 2. Exceptional Coxeter Arrangements
Supports 'E6', 'E7', 'E8', 'F4', 'G2', 'H3', 'H4'.

```
# F4 arrangement
F4 = exceptional_coxeter('F4')
print(f"F4: {F4.num_planes} planes in R^{F4.n}")
print("Characteristic Polynomial:", F4.characteristic_polynomial())
```

```
# H3 arrangement (Icosahedral symmetry)
# The base field is an algebraic number field containing the golden ratio.
H3 = exceptional_coxeter('H3')
print(f"H3: {H3.num_planes} planes in R^{H3.n}")
print("Base ring:", H3.K)
# H3 characteristic polynomial takes a moment due to NumberField arithmetic, 
# but works! (Uncomment to compute)
# print("Characteristic Polynomial:", H3.characteristic_polynomial())
```

## 3. Affine / Deformation Arrangements
These arrangements are naturally affine. The library returns their central cone (dimension + 1).

```
# Shi arrangement
S = shi(3)
print(f"Shi(3): {S.num_planes} planes in R^{S.n}")
print("Characteristic Polynomial:", S.characteristic_polynomial())
```

```
# Catalan arrangement
C = catalan(3)
print(f"Catalan(3): {C.num_planes} planes in R^{C.n}")
print("Characteristic Polynomial:", C.characteristic_polynomial())
```

```
# Linial arrangement
L = linial(3)
print(f"Linial(3): {L.num_planes} planes in R^{L.n}")
print("Characteristic Polynomial:", L.characteristic_polynomial())
```

```
# Ish arrangement
I = ish(3)
print(f"Ish(3): {I.num_planes} planes in R^{I.n}")
print("Characteristic Polynomial:", I.characteristic_polynomial())
```
